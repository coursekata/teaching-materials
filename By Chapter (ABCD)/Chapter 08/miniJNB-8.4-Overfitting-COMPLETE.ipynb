### 1.4 Which model of `mpg` is best? Why?

Is a model that  splits `wt` into 3 groups (**lighter** vs **medium** vs **heavier** vehicles) better than a model that just splits `wt` into 2 groups (**lighter** vs **heavier** vehicles). How about 4 groups? What happens to ***PRE*** and ***F*** as we modify the number of predictors in our model (i.e., as we increase the ***degrees of freedom*** used in our model)?

Run the two code boxes below to fit and visualize these models. Explain which model is the best fit, and why?

```r
# Run this code to fit the 2-group, 3-group, and 4-group models of 'mpg'

# 2-Group Model
mtcars$wt_2 <- ntile(mtcars$wt, 2)
mtcars$wt_2 <- factor(mtcars$wt_2)
model_2 <- lm(mpg ~ wt_2, data = mtcars)
supernova(model_2)

# 3-Group Model
mtcars$wt_3 <- ntile(mtcars$wt, 3)
mtcars$wt_3 <- factor(mtcars$wt_3)
model_3 <- lm(mpg ~ wt_3, data = mtcars)
supernova(model_3)

# 4-Group Model
mtcars$wt_4 <- ntile(mtcars$wt, 4)
mtcars$wt_4 <- factor(mtcars$wt_4)
model_4 <- lm(mpg ~ wt_4, data = mtcars)
supernova(model_4)
```

```r
# Run this code to visualize the 2-group, 3-group, and 4-group models of 'mpg'
model_2
gf_jitter(mpg ~ wt_2, data = mtcars)%>%
  gf_model(model_2)

model_3
gf_jitter(mpg ~ wt_3, data = mtcars)%>%
  gf_model(model_3)

model_4
gf_jitter(mpg ~ wt_4, data = mtcars)%>%
  gf_model(model_4)
```

```r
# OPTIONAL
# Run this code to visualize the residuals
gf_jitter(mpg ~ wt_2, data = mtcars)%>%
  gf_model(model_2)%>%
  gf_resid(model_2)

gf_jitter(mpg ~ wt_3, data = mtcars)%>%
  gf_model(model_3)%>%
  gf_resid(model_3)

gf_jitter(mpg ~ wt_4, data = mtcars)%>%
  gf_model(model_4)%>%
  gf_resid(model_4)
```

```r
# OPTIONAL
# Run this code to visualize the squared residuals
gf_jitter(mpg ~ wt_2, data = mtcars)%>%
  gf_model(model_2)%>%
  gf_squaresid(model_2)

gf_jitter(mpg ~ wt_3, data = mtcars)%>%
  gf_model(model_3)%>%
  gf_squaresid(model_3)

gf_jitter(mpg ~ wt_4, data = mtcars)%>%
  gf_model(model_4)%>%
  gf_squaresid(model_4)
```

<div class="teacher-note">

<b>Sample Response</b>: If we evaluate the models based on their F value, then the 2-group model is best, and the 4-group model is better than the 3-group model.
    
Note:
    
- Get students to think about why the F ratio goes down then back up as the predictors increase, and to connect that to SS Model and DF model, and/or SS Error and DF Error in the ANOVA table. 
- When looking at the visualizations, ask them what they notice about the way the distrubtion gets divided up based on number of groups, and how that changes the predictions for each group, and how the error for each of those model predictions connects back to the ANOVA table.   
- Optional: You can also show them the code for `gf_resid()` and `gf_squaresid()` to help them visualize the residuals and SS error.

## 2. Maximize your degrees of freedom

### 2.1 Make your own model.

### 2.2 What is the **maximum** number of groups you can add? 

Create a model with maximum complexity and compare your models to decide what the best model is.

```r
# sample response

# Create your models here

# 31-Group Model
mtcars$wt_31 <- ntile(mtcars$wt, 31)
mtcars$wt_31 <- factor(mtcars$wt_31)
model_31 <- lm(mpg ~ wt_31, data = mtcars)
model_31
supernova(model_31)


gf_jitter(mpg ~ wt_31, data = mtcars)%>%
  gf_model(model_31)
```

<div class="teacher-note">

<b>Sample Response</b>: We can only fit models with as many predictors as our degrees of freedom. A model with 31 groups has a PRE of 1 (100% explained variation) but the F value is very low. 
    
Note: Hopefully students will also start to question how practical a model like this might be. 
    
    
</div>

### 2.3 A highly complex model can closely match the data by using many degrees of freedom. What are some reasons a simpler model might still be more useful?

<div class="teacher-note">

<b>Sample Response</b>: A simpler model can be more useful because it is easier to understand, interpret, and apply to new data. While a complex model may fit the existing data very closely, it can “overfit,” meaning it captures random noise rather than the true underlying pattern. This makes it less reliable for predicting new outcomes. In contrast, a simpler model focuses on the most important relationships, which often makes it more generalizable and useful for explaining and predicting real-world situations. 
    
    
</div>
