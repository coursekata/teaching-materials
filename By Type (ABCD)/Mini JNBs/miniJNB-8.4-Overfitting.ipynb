### 1.4 Which model of `mpg` is best? Why?

Is a model that  splits `wt` into 3 groups (**lighter** vs **medium** vs **heavier** vehicles) better than a model that just splits `wt` into 2 groups (**lighter** vs **heavier** vehicles). How about 4 groups? What happens to ***PRE*** and ***F*** as we modify the number of predictors in our model (i.e., as we increase the ***degrees of freedom*** used in our model)?

Run the two code boxes below to fit and visualize these models. Explain which model is the best fit, and why?

```r
# Run this code to fit the 2-group, 3-group, and 4-group models of 'mpg'

# 2-Group Model
mtcars$wt_2 <- ntile(mtcars$wt, 2)
mtcars$wt_2 <- factor(mtcars$wt_2)
model_2 <- lm(mpg ~ wt_2, data = mtcars)
supernova(model_2)

# 3-Group Model
mtcars$wt_3 <- ntile(mtcars$wt, 3)
mtcars$wt_3 <- factor(mtcars$wt_3)
model_3 <- lm(mpg ~ wt_3, data = mtcars)
supernova(model_3)

# 4-Group Model
mtcars$wt_4 <- ntile(mtcars$wt, 4)
mtcars$wt_4 <- factor(mtcars$wt_4)
model_4 <- lm(mpg ~ wt_4, data = mtcars)
supernova(model_4)
```

```r
# Run this code to visualize the 2-group, 3-group, and 4-group models of 'mpg'
model_2
gf_jitter(mpg ~ wt_2, data = mtcars)%>%
  gf_model(model_2)

model_3
gf_jitter(mpg ~ wt_3, data = mtcars)%>%
  gf_model(model_3)

model_4
gf_jitter(mpg ~ wt_4, data = mtcars)%>%
  gf_model(model_4)
```

```r
# OPTIONAL
# Run this code to visualize the residuals
gf_jitter(mpg ~ wt_2, data = mtcars)%>%
  gf_model(model_2)%>%
  gf_resid(model_2)

gf_jitter(mpg ~ wt_3, data = mtcars)%>%
  gf_model(model_3)%>%
  gf_resid(model_3)

gf_jitter(mpg ~ wt_4, data = mtcars)%>%
  gf_model(model_4)%>%
  gf_resid(model_4)
```

```r
# OPTIONAL
# Run this code to visualize the squared residuals
gf_jitter(mpg ~ wt_2, data = mtcars)%>%
  gf_model(model_2)%>%
  gf_squaresid(model_2)

gf_jitter(mpg ~ wt_3, data = mtcars)%>%
  gf_model(model_3)%>%
  gf_squaresid(model_3)

gf_jitter(mpg ~ wt_4, data = mtcars)%>%
  gf_model(model_4)%>%
  gf_squaresid(model_4)
```
