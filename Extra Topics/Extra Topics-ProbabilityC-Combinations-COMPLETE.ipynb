## 3.0: Comparing the Random DGP to the REAL DGP

**3.1:** In the real story, the three sailors who had antibodies did not contract COVID19. 
Is that unlikely to happen from a random DGP? (How would we define "unlikely"?)

```r
# feel free to play around with this code to explore the random DGP
do(10) * sum(sample(antibodies,100))
```

```r
# sample response
samplingdistribution <- do(1000) * sum(sample(antibodies,100))
tally(~ sum, data = samplingdistribution, format = "proportion")

# ggformula plot that does not work in datacamp but works in jupyter
#gf_props(~ sum, data = samplingdistribution)
```

<div class="alert alert-block alert-warning">

**Sample Response**

In our simulations, about 50% of the the time, all three sailors with antibodies get COVID19.
It's quite rare for 0 of the sailors with antibodies to get COVID19 (about 5 in a 1000)!
