### 2.0 - Estimating the DGP with our Sample

#### $p$ vs  $\hat p$:

We can use the following notation to distinguish $p$ from $\hat p$:

$p$ = population proportion
> - parameter
> - the true DGP

$\hat p$ = sample proportion
> - statistic
> - estimator of $p$

**What is their statistical claim?**

So, how do we determine how far off our estimates will usually be? We use the sampling distribution!

### 3.0 - Sampling Distribution of Proportions

We know that, under certain conditions, the sampling distribution of proportions will be approximately normal ($\hat p$ ~ Normal), centered at the true population proportion ($\mu_{\hat p} = p$), with some variation around that center ($\sigma_{\hat p} = \sqrt \frac {p (1-p)}{n}$).

<img src="https://i.postimg.cc/7wYG25Qp/CI-for-Prop-SDo-P-conditions.png" title="Sampling Distribution of Proportions conditions" width = 800/>

```r
# COMPLETE VERSION

# just some code to show you how to generate samples (n = 30)
percent_men <- 60
population <- c(rep(0,100-percent_men), rep(1,percent_men))


samp_dist_of_proportions <- do(1000) * (sum(resample(population, 30)) / 30)

gf_histogram(~ result, data = samp_dist_of_proportions)
```

So if we plug in what we know and do all the calculations, we can generate a sampling distribution of proportions, as seen below, that results in a center, $\mu_{\hat p}$, at 60%, and a standard error, $\sigma_{\hat p}$, of about 8.9%.

<img src="https://i.postimg.cc/TfF2Q09v/CI-for-Prop-SDo-Prop.png" title="Sampling Distribution of Proportions" width = 600/>

So with these parameter estimates, based on $Normal(\mu_{\hat p} = 0.6, \sigma_{\hat p} = .089)$, we can see that our sample proportion ($\hat p$) will typically deviate from the true DGP ($p$) by about 8.9%.

### 4. 0 - Confidence Intervals

There should be a way to report both:

> a) Our estimate

> b) How far off this estimate might be

One way to do this is with a confidence interval!

**The EU Migration Study – Spain**

In order to construct a 95% confidence interval of proportions, we will use data from Spain, which has the best records on EU migrants (both documented and undocumented migrants). According to a report\* in Comparative Migration Studies, Spain has a “nationwide rolling local population register, which provides a comprehensive register of the population, also including unauthorized immigrants.” The Immigrant Citizens Survey** conducted a random sample of this database, surveying recent immigrants to Spain who were 15 and older.

***Sources***

*\*[Reichel and Morales](https://www.ncbi.nlm.nih.gov/pmc/articles/PMC5209407/pdf/40878_2016_Article_44.pdf), “Surveying immigrants without sampling frames – evaluating the success of alternative field methods,” Comparative Migration Studies (2017) 5:1* 

*\**Huddleston, Thomas and Dag Tjaden, Jasper. ”Immigrant Citizens Survey: How immigrants experience integration in 15 European cities,” King Baudouin Foundation and The Migration Policy Group. May 2012.*

We’ll first look at mock data from this survey, for the purposes of this lesson, then we will look at the real data.

**Scenario (Mock Data):** 

In reality, we don’t know the true proportion of recent immigrants to Madrid
(Spain’s immigration hub) who are male because it’s almost impossible to collect records about each
individual. So, instead, imagine we surveyed a random sample of 34 immigrants in Madrid’s address
registries. Of the random sample, 15 identified as male. 

```r
# Use the sample to estimate the proportion of migrants who are male
15 / 34
```

Once again, we have this sample estimate to estimate the proportion in the DGP, but we know samples can be wrong. So instead of a single estimate, we are going to produce a range of estimates: a confidence interval.

***Your goal:***

**Estimate** the true proportion of Madrid’s recent immigrants who identify as male, with **95% confidence**.

#### A DGP where $p = \hat p$

We are going to put away any claims and just look at our sample proportion ($\hat p$) and see: How confident are we that we get a certain estimate around that sample proportion?

What is the true proportion of recent Madrid immigrants who are male? Well, we don't know ($p = ?$), but we *do* have our sample proportion ($\hat p = 15/34 = 0.44$).

So we can say 44% of the people in our sample were males, but how much error should we give this sample estimate? What are the boundaries that give us 95% of the most typical estimates?


<img src="https://i.postimg.cc/S45wTftm/CI-for-Prop-SE-for-p-hat.png" title="What is SE for p-hat" width = 600/>

To help us figure this out, we will need to consider all the possible random samples that could have happened (a sampling distribution!).

#### Since we don't know $p$ and we don't know $\sigma$:

<img src="https://i.postimg.cc/5xrBw3qr/CI-for-Prop-Unknown-p.png" title="We do not know true proportion" width = 600/>

#### We can use our sample, $\hat p$, and standard error $SE_{\hat p}$:

<img src="https://i.postimg.cc/RvVKy8g8/CI-for-Prop-Use-sample-p-hat.png" title="We can use sample proportion" width = 600/>

#### Plug in what we know ($\hat p$, $n$) and calculate:

<img src="https://i.postimg.cc/Dhggjrf8/CI-for-Prop-Calc-Sample-Prop.png" title="Calculations for Sample Estimates" width = 600/>

#### We get the resulting sampling distribution with the following estimates: 

#### $\hat p=44\%$ and $SE_{\hat p}=8.5\%$

<img src="https://i.postimg.cc/RSyqspNk/CI-for-Prop-Sample-Prop-Estimates.png" title="Proportions from Sample Estimates" width = 600/>

Now, if you will recall from the empirical rule, in order to capture the middle 95% of cases we need to go about 2 $\sigma$s above and below the mean. To be more precise, we need to go 1.96 $\sigma$s out from the mean. 

This can be summarized with the formula below (substituting $\sigma$ fwith $SE_{\hat p}$).

**How to find a Confidence Interval:**

$\hat p +/- 1.96*SE_{\hat p}$

```r
# Find the margin of error (1.96*SE)
1.96 * .085
```

With this margin of error, we know that if we go 0.167 above and 0.167 below the center proportion (0.44), this will give us our lower and upper boundary for the middle 95% of sample proportions.

```r
# Lower Bound
0.441 - 0.167

# Upper Bound
0.441 + 0.167
```

Now we have our 95% confidence interval, and we can say with 95% confidence that the true proportion of migrants who are male is between 27.% and 60.8%.

Careful not to say there is a 95% "probability", because that is not the case, probability is not the same as a confidence interval.
