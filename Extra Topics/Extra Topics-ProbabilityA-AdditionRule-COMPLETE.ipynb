### 2.0 - Ways to Express Probability

**Probability Notation**

Since we are trying to figure out the probability of some event, we will need some ways to put our words into math language. 

For instance, we can say *"the probabilty of Event X"* like this: 

$P(X)$

In this case, we may want to figure out *"the probability of being tall"*, so, in math terms, we could say it this way: 

$P(Tall)$

**2.1:** How might we express *"The probability of being wealthy"*?

<div class="alert alert-block alert-warning">

**Sample Response**

</div>

We can also expand that notation to make statements about the probability of being tall AND wealthy, or being tall OR wealthy.

These are mathematical ways to say these words:

- $\cup$ is the symbol for **Union** and is a way to say **"Or"**
- $\cap$ is the symbol for **Intersection** and is a way to say **"And"**

**2.2:** For example, which of these says *"the probability of being tall and wealthy"*? What does the other expression say?

> **A.** $P(Tall \cap Wealthy)$

> **B.** $P(Tall \cup Wealthy)$

<div class="alert alert-block alert-warning">

> **A.** $P(Tall \cap Wealthy)$

The probability of being tall OR wealthy:

> **B.** $P(Tall \cup Wealthy)$

### 5.0 - Estimating $P(T \cup H)$ and $P(T \cap H)$ 

**INTERSECTION ($\cap$): P(T "AND" H)**

If our goal is to estimate the probability of finding someone who is taller AND wealthier than the median--**P(T $\cap$ H)**--it can be helpful to think of the $\cap$ to mean **"and"** or **"both"**. This means they have to be part of *both* the tall *and* the wealthy group (the intersection of the two categories). 

We can visualize this idea in a Venn diagram or a table. 

If we think of this as the entire event space (all observed outcomes):

<img src="https://i.postimg.cc/6Bh3qDPB/Addition-Rule-Event-Space.png" title="Event Space" width = 700/>

Then, we can count the observations for the space (or spaces) we are trying to estimate. In this case, we are interested in the intersection of the "Tall" and "High-Earner" groups. There is only one space that meets that condition:



<img src="https://i.postimg.cc/kCf2x1mT/Addition-Rule-P-of-T-and-H-Diagrams.png" title="Probability of T and H" width = 600/>

```r
# Find the probability by dividing by N
101 / 192
```

We can also use the `&` symbol in R to help us estimate this intersection!

```r
# Intersection = &
tally(~ (height2=="Tall" & income2=="High-Earner"), data = profiles)

# Divide by N to get the probability
tally(~ (height2=="Tall" & income2=="High-Earner"), data = profiles) / 192
```

**5.1:** Is this probability close to what we would expect if all the profile creators were reporting their information accurately?

<div class="alert alert-block alert-warning">

**Sample Response**

No, this probability (about 53%) is much higher than we would expect under the null model (which predicts 25%).

</div>

**UNION ($\cup$) = P(T "OR" H)** 

If we want to estimate the probability of finding someone who is tall, OR wealthy, OR both tall and wealthy, then that means we ***add*** all those possibilities together.

**Take Note:**

- $\cup$ = "OR" = "one, the other, or both"
- $\cup$ = "add" = P(T) + P(H) - P(T$\cap$ H)

In other words, below we can see that P(T) = 22 + 101, and P(H) = 48 + 101. Since we don't want to double count the intersection (101), we would just subtract that, or be careful not to double count it!

<img src="https://i.postimg.cc/X4cSChMJ/Addition-Rule-P-of-T-or-H-Diagrams.png" title="Probability of T or H" width = 600/>

```r
# Add the counts then divide by N to get the probability
(22 + 101 + 48) / 192
```

We can also use the `|` symbol in R to help us estimate this union!

*NOTE: it is the vertical line symbol--not a lower case "L"--and is usually paired with the "backslash" (\\) key near the "backspace" or "delete" key on your keyboard.*

```r
# Union = |
tally(~ (height2=="Tall" | income2=="High-Earner"), data = profiles)

# Divide by N to get it as a proportion
tally(~ (height2=="Tall" | income2=="High-Earner"), data = profiles) / 192
```
