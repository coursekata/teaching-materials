The dataframe is called `GullStare` and is preloaded into this jupyter notebook.

Here is how the researchers collected their data.

- 250 g of fried potato chips in front of the experimenter. 
The food was presented inside a sealed, transparent freezer bag 
(to prevent gulls from eating the food).
- Each gull was given two chances (called trials) to approach the food: 
one trial where the experimenter looked directly at the gull and the 
other trial where the experimenter looked away. 
- Half the gulls got the **Looked At** trial first and the other half got 
the **Looked Away** trial first.
- If a gull did not touch the food within 300 s of starting its approach 
(e.g., took a step toward the food) but remained in the vicinity, the trial 
was deemed complete and a time of 300 s was recorded. 
- If the gull approached the food quickly, they have a shorter latency (e.g.,
60 s). If they took a longer time to approach the food, they have a longer
latency (e.g., 200 s).

For more details about this experiment, check out [the full article](https://royalsocietypublishing.org/doi/10.1098/rsbl.2019.0405) 
published in **Biology Letters**.

## 3.0 - The Superpower of Simulations

Now I want you to imagine that the world, “the population/DGP,” is just like our sample. 
Imagine doing this experiment again where we randomly select a slightly different 
group of 67 gulls and calculated a slightly different estimate of $\beta_0$. 

We're going to use two methods to mimic a world that looks like our sample data:

1. Bootstrapping (`resample()`)

2. Simulating from a normal distribution (`rnorm()`)

We're going to try and figure out: if the world was basically like our sample,
how would our estimate of $\beta_0$ vary? What kind of estimates could we get?

## 4.0 - Bootstrapping the SDoM

4.1 - What is the difference between the first and second line of code below?
How does the 2nd line of code mimic a sample that is "similar to our data"?

```r
mean(GullStare$Longer)

mean(resample(GullStare, 67)$Longer)
```

<div class="alert alert-block alert-warning">

    
**Sample Response**
    
- the first one is just the sample mean ($b_0$)

- the second one -- notice the data is the resampled data... so it always changes slightly
 

**Note to Instructors**

Resample simulates some sampling variation. If your students are curious, you may want to clarify that the idea behind resampling is that it simulates a population that is made of many copies of our data. Like if your data is A A B D, then the population that's like your data is A A B D, A A B D, A A B D, A A B D, A A B D, A A B D, etc. And sampling from this large population you can get samples like A B B D or occasionally D D D D, just like with resample.

</div>

4.2 - Add the function `do()` to generate a 1000 $b_0$s. Save your **S**ampling
**D**istribution **O**f **M**eans as **sdom_resample**.

```r
# edit this code
mean(resample(GullStare, 67)$Longer)
```

```r
# COMPLETE
sdom_resample <- do(1000) * mean(resample(GullStare, 67)$Longer)
```

6.2 - Run the code below. Note that the means and standard errors of these

```r
favstats(~ mean, data = sdom_resample)
favstats(~ mean, data = sdom_rnorm)
```

<div class="alert alert-block alert-warning">

**Sample Response**
    
- both processes are like drawing from a distribution with the same mean and sd (the same ones from the sample)! so you get sampling distributions that are very similar (even though the individuals you sample might be very different, the resulting $b_0$s -- the means -- are very similar!)
