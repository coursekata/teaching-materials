### 1.0 - The Flint Water Crisis

In 2014, city officials in Flint, Michigan, celebrated the cost-cutting measure of switching the city's water source from Detroit to the local Flint River. By 2015, residents of Flint began to report problems with the water (e.g., discolored water, foul smelling water, health problems) and complained that the water was unsafe. Researchers claimed that the new water from the Flint River corrodes pipes, so lead gets into the water. Among many other side effects, ingesting lead can hurt children’s brain development. ### 4.0 - How unlikely is this data?

We can see that our sample proportion falls in the upper tail of the distribution. If we use the Empirical Rule (the 68 – 95 – 99.7 Rule), it suggests that it is pretty unlikely (less than 5% chance) to get a sample proportion more than two standard deviations out from the center.

How do we figure out how many standard deviations away from the center our sample is?

#### From Z-Scores to Z-Test Statistics

A z-score tells us how many standard deviations a value is above or below the mean (or the empty model).

$z = \frac {data value - mean}{standard deviation}$

We can extend this idea to our sample statistics andjust take the distance between the sample proprtion and the empty model and divide it by the standard deviation (or standard error) to see how many standard deviations fit in that distance.

$z = \frac {\hat p - \mu_{\hat p}}{\sigma_{\hat p}}$

##### How many standard deviations away from $p_0$ is $\hat p$?

<img src="https://i.postimg.cc/hDn0TY6c/Ht-1-Prop-z-score.png" title="Z Score" width = 800/>

If 2 standard deviations away from $p_0$ would be unlikely, then 3.5 standard deviations away is *very* unlikely.

The larger the test statistic (in magnitude), the more unusual your measurements are under the null
assumption. We should probably start to doubt the $p_0$ assumption.

But exactly how unlikely was the proportion we sampled?

For this, we need p-value!

<img src="https://i.postimg.cc/CSshCr1J/HT-1-Prop-Find-this-Probability.png" title="Find the Probability" width = 800/>

```r
#find p-value
pnorm(q=3.53, lower.tail=FALSE)
```
