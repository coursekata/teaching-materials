### 2.0: Assumptions and ANOVA

Whenever we conduct an ANOVA (i.e., ANalysis Of VAriance) we are making some assumptions. One of those assumptions is that the groups we are comparing have roughly equal variances. This is usually not something to worry about when the groups are normally distributed, and are close in sample size, but if they are not, we should take an extra step and confirm whether our groups have equal variances (also called homogeneity of variances, or Levene's test) before proceeding with any ANOVA.

**Discussion:** Why do you think it matters whether the groups we are comparing have roughly equal variances?

<div class="alert alert-block alert-warning">

**Sample Responses**

- So we don't draw incorrect conclusions when interpreting out p-value
- To avoid Type I and/or Type II Error
- Because a group with very little variance might be quite different from a group with a lot of variance

</div>

**2.1:** Take a look again at the distribution of the two `respect_condition` groups. What is the shape of their distributions? Would you guess that they have roughly equal variances (or spread)?

**2.2:** We can easily check this assumption with the function `leveneTest()`, but before we conduct a hypothesis test for variance, let's figure out what the empty model assumes in this analysis.

The empty model assumes which of these?

A. All variances are equal

B. At least two of them are different

<div class="alert alert-block alert-warning">

**Sample Response**

</div>

**2.3:** Run the analysis below. Considering the answer to the previous question, how do we interpret the p-value?

```r
# Load library(car) (needed for leveneTest() function)
library(car)


leveneTest(veg_eaten ~ respect_condition, data = respect_study)
```

<div class="alert alert-block alert-warning">

**Sample Response**

We would interpret the p-value of 0.83 to mean that it is very likely (83% probability) that the variances are roughly equal (i.e., they came from a DGP where the empty model is true)

This means that there is no evidence to suggest that the variance in `veg_eaten` is statistically significantly different for the two respect condition groups.

</div>

**2.4:** What does this mean for our overall analysis?

<div class="alert alert-block alert-warning">

**Sample Response**

It means that we will not be violating the assumption of equal variances among groups when we conduct our ANOVA. This will also ensure that we can make valid interpretation of our ANOVA results.

</div>
