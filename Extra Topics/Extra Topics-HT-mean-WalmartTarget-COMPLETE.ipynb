Everyday prices are changing at Walmart and Target. Everyday they are trying to beat each other.
There are also so many more products at each of these stores than what is included in this small data set
of 174 things.

What we want to know is not just about the price differences of these particular items, but the
DGP that produced these price differences: What is it like? That's the question we will be focusing on.

3.1 - Is it possible that the two stores do NOT really differ in their prices?
Could that general idea be true even if some items are more expensive at Walmart and
some items are more expensive at Target?

3.2 - If we assumed that the two stores do not really differ in price, what should we 
expect the average difference between the prices to be? 

If this was true, what would be the best $\beta_0$ to represent the Data 
Generating Process? Write that number in the GLM below.

$Y_i = \beta_0 + \epsilon_i$

<div class="alert alert-block alert-warning">

We would expect the average price difference to be zero. 

$Y_i = 0 + \epsilon_i$

We are using $\beta_0$ and $\epsilon_i$ because we are estimating the population, or DGP.

## 4.0 - Modeling Variation

4.1 - We cannot ever really know $\beta_0$. But we can at least estimate it. It's better
than guessing, right? What is our best estimate of the DGP from our sample?

$Y_i = \beta_0 + \epsilon_i$

```r
# Get the estimate for the average price difference with lm()
lm(PriceDiff ~ NULL, data = Shopping)

# Or get the estimate with b0() 
b0(PriceDiff ~ NULL, data = Shopping)

# Or get the estimate with mean() 
mean(~ PriceDiff, data = Shopping)
```

<div class="alert alert-block alert-warning">

**Sample Response**

Our best estimate is 0.08.

$Y_i = 0.08 + \epsilon_i$
    
</div>

## 5.0 - Simulating the DGP

Since statistics and data science is all about *what's possible*, let's consider this:

**We are trying to imagine that the world of all the stuff you could
buy at Target or Walmart (“the population”) is similar to the price differences
reflected in our sample of 174 items. How much could our $b_0$ vary if we had taken a 
bunch of different samples, over and over again?**

5.1 - Run the code below a few times and observe how it works. Why does it create 174 numbers? 


```r
PriceDiff_stats <- favstats(~ PriceDiff, data = Shopping)
rnorm(174, PriceDiff_stats$mean, PriceDiff_stats$sd)
```

<div class="alert alert-block alert-warning">

**Sample Response**

`rnorm()` is a data generating process that creates a sample of 174 numbers from a population that has the same mean and standard deviation as `PriceDiff`. You can think of this code as generating a new sample of 174 (fairly similar to our empirical sample) every time it is run.
</div>

5.2 - Will the code below also generate 174 numbers? Why or why not?

```r
mean(rnorm(174, PriceDiff_stats$mean, PriceDiff_stats$sd))
```

<div class="alert alert-block alert-warning">

**Sample Response**

This code takes a sample of 174 values and calculates the mean; thus it only generates 1 number.

</div>

5.3 - The code below will generate 10 $b_0$s. 

What do you notice about the generated $b_0$s?
What are they typically like? How big can they get? How small can they get?
If you generated a 1000 of these $b_0$s, what would the resulting distribution look like?

```r
do(10) * mean(rnorm(174, PriceDiff_stats$mean, PriceDiff_stats$sd))
```

<div class="alert alert-block alert-warning">

**Sample Response**

They all seem to range within less than 1 dollar above or below zero. If we had 1,000 of these 
estimates the distribution would probably be centered around zero and spread around it 
within about 1 dollar above or below that.
    
</div>

```r
# COMPLETE
# Modify to simulate 1000 samples, and save it into an object called sdob
sdob <- do(1000) * mean(rnorm(174, PriceDiff_stats$mean, PriceDiff_stats$sd))

# This gives us a peek at what is inside sdob
head(sdob)

# Create a visualization of sdob
gf_histogram(~ mean, data = sdob)
```

5.5 - This is a sampling distribution created from the assumption that the DGP is basically like 
our sample. What is the DGP represented by `rnorm(174, PriceDiff_stats$mean, PriceDiff_stats$sd)`?
Put it in the GLM equation below.

$Y_i = \beta_0 + \epsilon_i$

<div class="alert alert-block alert-warning">

**Sample Response**

The DGP represented by that code has the same mean as our sample. Note the `PriceDiff_stats$mean` in our code:

`rnorm(174, PriceDiff_stats$mean, PriceDiff_stats$sd)`

Thus, the DGP represented by this code can be written as:

$Y_i = 0.08 + \epsilon_i$
    
</div>

5.6 -In the histogram below we have colored in the middle 95% of means to show that they are "not unlikely" to come
from this DGP. **Is our sample mean (it's going to appear as a vertical line) going to be in the "not unlikely" zone?**

Make a prediction, then run the code below.

```r
sdob <- do(1000) * mean(rnorm(174, PriceDiff_stats$mean, PriceDiff_stats$sd))

# This will create a visualization that includes our empty model
gf_histogram(~ mean, data = sdob, binwidth = .05, fill = ~middle(mean, .95)) %>%
  gf_vline(xintercept = ~PriceDiff_stats$mean)
```

<div class="alert alert-block alert-warning">

**Sample Response**

Our sample $b_0$ estimate is within the "not unlikely" zone. This makes a lot of sense because the population mean is **just** like our sample mean.
    
</div>
