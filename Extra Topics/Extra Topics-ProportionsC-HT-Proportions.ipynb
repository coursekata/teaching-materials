### 1.0 - The Race/Resumé Study

In the United States we have seen a persistent racial wage gap. Specifically, there is a significant difference between the median earnings of white people and black people.

<img src="https://i.postimg.cc/mL0x25Lv/HT-2-Prop-Earnings.png" title="Earnings" width = 600/>

[Economic Policy Institute, 2018](https://www.epi.org/blog/black-workers-have-made-no-progress-in-closing-earnings-gaps-with-white-men-since-2000/)

We may be able to at least partially explain this racial wage gap with social factors, such as inequity of schools, differences in the social networks people have access to, or differences in choice of career. On the other hand, there may still be direct discrimination in the labor market during the hiring process.

Researchers wanted to test if direct hiring discrimination was still a factor in labor markets. In their study*, investigators created mock resumés that were identical except that each resumé was randomly assigned either a commonly-white or commonly-black name. 

<img src="https://i.postimg.cc/q4X1qvpJ/HT-2-Prop-Resumes.png" title="Resumes" width = 600/>

Note: These names were used as the main example in the original study paper.

The resumés were sent to 4,890 job placement ads in Chicago and Boston, and included a wide swath of jobs in the following industries: sales, administrative support, clerical services, and customer services. They also included a large range of positions, from “cashier work at retail establishments and clerical work in a mailroom to office and sales management positions.” The researchers measured which group got more callbacks from potential employers.

Note: This lesson presents a slightly simplified version of their study design. However, the data in the lesson is identical to the real study data. See citation below for their full paper:

*\*[Bertrand, Marianne and Sendhil Mullainathan](https://www.nber.org/papers/w9873). "Are Emily And Greg More Employable Than Lakisha And Jamal? A Field Experiment On Labor Market Discrimination," American Economic Review, 2004, v94(4,Sep), 991-1013.*


**Today’s Key Analysis**

Does a statistically significant difference in callbacks indicate broader discriminatory practices?

### 3.0 - A DGP where $p_1 - p_2 =0$

Are these proportions **different enough** to show discrimination, or could this difference have been a
result of **chance alone**?

Let's see if we can statistically rule out the random model.

First, let's assume a DGP where the empty model is true, and there is no discrimination; these resumés are getting callbacks by random chance.

##### Assume the Empty Model is true:

<img src="https://i.postimg.cc/RMQqNjmY/HT-2-Prop-Empty-Model.png" title="Assume Empty Model is True" width = 600/>

Next, we ask ourselves: if the empty model is in fact true, how likely is the data that we've gathered?

The study sent resumés to 4890 businesses (n = 2445 each group). At this sample size, what estimates for $p_1 - p_2$ are we likely to get if there’s actually no discrimination? Let’s find the randomization distribution of $\hat p_1 - \hat p_2$ from all possible random assignments.

Under certain conditions, we know that the distribution will be approximately normal, and centered at zero ($\mu = 0$) because we are assuming a DGP where there is no difference in proportion of callbacks for each group, with some amount of variation ($\sigma$).

##### A Sampling Distribution of Differences

<img src="https://i.postimg.cc/9C9D6x0r/HT-2-Prop-Samp-Dist-of-P.png" title="Sampling Distribution under certain conditions" width = 600/>

Note that, in the formula for $\sigma$, the $\hat p_c$ represents the combined proportion of the two groups.

```r
# P1 
246 / 2445

# P2
164 / 2445

# Pc
410 / 4890
```

This gives us the calculations below for the standard error of the distribution.

<img src="https://i.postimg.cc/dQY1RCP4/HT-2-Prop-Sigma-Formula-Calc.png" title="Sigma calculations" width = 600/>


Resulting in a sampling distribution with the following parameters: $\mu = 0, \sigma = 0.0079$.

<img src="https://i.postimg.cc/ChVY9VPK/HT-2-Prop-Samp-Dist-Parameters.png" title="Sampling Distribution Parameters" width = 600/>

**3.2:** But just to be sure, let's get a precise p-value. We can use the function `prop.test()` again to get this and other helpful information. 

```r
# Run the analysis to get the p-value
prop.test(x = c(246, 164), n = c(2445, 2445))
```

```r
# COMPLETE VERSION

# Note to teachers:
# default is two tailed test 
prop.test(x = c(246, 164), n = c(2445, 2445))

# if you want to test whether first group is greater than second group
prop.test(x = c(246, 164), n = c(2445, 2445), alternative = "greater")

# if you want to test whether first group is less than second group
prop.test(x = c(246, 164), n = c(2445, 2445), alternative = "less")

# Alternatively, you can use pnorm()
# Find the z-score
(.034 - 0) /.0079

#find p-value
# two-tailed 
pnorm(q=4.30, lower.tail=TRUE)
# one-tailed
pnorm(q=4.30, lower.tail=FALSE)
```

**5.1: Discussion Question:** 

We found that commonly-white name resumés had a statistically significant advantage in attracting callbacks. Does this result suggest discrimination at other levels of the hiring process (e.g. job offers rates after an interview, starting salary estimates, etc.)? Explain.

<div class="alert alert-block alert-warning">

**Sample Response**

Since the only outcome measured in this study was callback rates for resumés, we cannot generalize this
study result to other parts of the hiring process. For example, the study did not investigate any potential
differences in starting salary offers, employer responses to in-person interviews, or whether or not
applicants are given job offers. Every level of the hiring process may operate differently, therefore, the
discrimination found in call back rates does not necessarily indicate discrimination for the other levels.

However, if there is discrimination in callbacks, one could make the case that this initial roadblock would
signify a practically important challenge to the entire application process. It doesn’t matter if the
interview process, job offering process, and starting salary offer process is equal if applicants can’t even get to those parts of the process due to initial discrimination in the first place (no callback). In other words, if an applicant can’t get their foot in their door, it doesn’t matter how they’d be treated when they actually enter the room.

Another challenge to this study’s scope of inference comes from the fact that these resumés were sent
in response to newspaper ads for jobs. This is an uncommon channel for finding a job in today’s world.
Discrimination in more modern channels of applying for jobs could be better or worse – more study is
needed in this area.

</div>
