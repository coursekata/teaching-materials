### 2.0: Successive Independent Events

Let's build on some notions of probability; namely, the probability of ***successive independent events***.

**Successive:** Events are successive if one happens after the other (i.e. a repeated process)

**Independent:** Events are independent if knowing the outcome of one event doesn’t change the probability of another event.

Let's look at some examples.

**Example 1:** *What is the probability of getting “heads” 5x in a row?*

Recall our tree diagrams:

#### 1st Flip:

#### 2nd Flip: 
<img src="https://i.postimg.cc/1P84SDgG/Perm-and-Comb-2nd-Flip-H-or-T.png" title="Branch of probability of heads or tails then heads or tails again" width = 300/>
  

#### 5th Flip: 
<img src="https://i.postimg.cc/zJdWt5M8/Perm-and-Comb-5th-Flip-H-or-T.png" title="Branch of probability of heads 5 times in a row" width = 500/>
 

We can express this as P(5 heads in a row) = P(H∩H∩H∩H∩H)

**2.1:** Find the probability of getting heads 5 times in a row, or P(H∩H∩H∩H∩H).

**Remember:** “and” means multiply!

<div class="alert alert-block alert-warning">

**Sample Response**
 
P(5 heads in a row) = .5 ∗ .5 ∗ .5 ∗ .5 ∗ .5

P(5 heads in a row) = (.5)^5 = 0. 03

***Since the probabilities don’t change, you don’t need a tree diagram!***

**Why it makes sense to treat free throws as independent events:**

- No one is guarding the player

### 3.0 - "At Least One"

"At Least One" Scenarios:

When asked to find the probability of “at least one” occurrence of a successive independent event, take the **complement** of the event that **none** occur.

P(at least one) = 1 – P(none)

**Example:** 

Stephen Curry makes 90.6% of the free throws he attempts. Assume each of his free throw attempts are independent. What is the probability he makes at least one of his next 4 free throws?

We can represent this idea like this:

P(makes at least one) = P(make one or more)

So, we can see this probability includes several events together (the combinations of making one or more free throws). Thus, he could make all 4 shots (.906^4), OR (remember, "or" means "add") he could have made the first 3 and missed the last one [.906 ∗ .906 ∗ .906 ∗ .094], OR he could have made the first 2 and the fourth but missed the third [.906 ∗ .906 ∗ .094 ∗ .906]...and so on with the many possible outcomes. This obviously becomes an overly complicated set of calculations:

> P(makes at least one) = P(make one or more) = (.906)^4 + [.906 ∗ .906 ∗ .906 ∗ .094] +
> [.906 ∗ .906 ∗ .906 ∗ .094] + [.906 ∗ .906 ∗ .094 ∗ .906] + [.906 ∗ .094 ∗ .906 ∗ .906] + [.094 ∗ .906 ∗ > .906 ∗ .906] + [.906 ∗ .906 ∗ .094 ∗ .094] + ...and so on....

BUT... there is a simpler way!

We can think about it like this: The P(makes at least one) is just the opposite, or the *complement*, of P(makes none).

**P(makes at least one) <-- complements --> P(makes none)**

The only way you can't make "at least one" free throw is if you make none all 4 times!

Another way of saying this is:

P(makes at least one) = 1 – P(makes none)

**3.1:** What is the probability he makes at least one of his next 4 free throws?

<div class="alert alert-block alert-warning">

**Sample Response**

P(makes at least one) = 1 – .094 ∗ .094 ∗ .094 ∗ .094

OR

P(makes at least one) = 1 – .094^4 = 0. 99992 or 99.992%

</div>

### 4.0 - Sampling Without Replacement

Let's think of a more *dependent* situation: Sampling Without Replacement

Example: You have a jar with 12 blue marbles and 8 red marbles. Imagine you sample marbles without replacement. What is the probability of drawing the following: blue, then red, then blue.

P(blue, red, blue) = 12/20

“and” means multiply!

P(blue, red, blue) = 12/20 ∗ 8/19 ∗ 11/18 = 0. 154

#### Testing Kobe's "Hot Hand"

Special thanks to Matt DiSorbo for help with data mining and analysis for this section.

The “hot hand” theory: When a player starts to make many shots in a row, the have a “hot hand” – their probability of making shots is higher than normal.

<img src="https://i.postimg.cc/DnV1tTQF/Perm-and-Comb-Hot-Hands.png " title="Basketball with hands on fire" width = 300/>

This would mean your shots are not independent. Your current shot probability *depends* on your previous shots.

If anyone has had a “hot hand,” it was Kobe during his 81 point game.

Over his career, he was a 44.7% accuracy shooter. In the 3rd quarter of the game, he made 8 consecutive shots (not free throws). That game, was his probability of making shots higher than usual?


#### What would be the Empty Model?

Let’s assume the “hot hand” doesn’t exist, so each shot has an equal probability of going in, no matter a player’s previous results in the game. We assume each shot is independent.

**4.1:** What is the probability of Kobe making 8 shots in a row?

<div class="alert alert-block alert-warning">

**Sample Response**

P(make 8 in a row) = (.447)^8 = 0.00159 (or 0.159%)

OR

P(make 8 in a row) = 0.447 ∗ 0.447 ∗ 0.447 * 0.447 ∗ 0.447 ∗ 0.447 * 0.447 ∗ 0.447 = 0.00159

</div>

**4.2:** Does this mean it's likely he made that streak of shots by chance, or that it is likely he had a "hot hand" that day?

<div class="alert alert-block alert-warning">

**Sample Response**

This is a very low probability that he made them by chance. It seems he likely had a "hot hand".

</div>

But Kobe took a lot of shots...In his career, Kobe took:

30,697 shots*

How many 8-shot streaks would we expect?

\*Note: *Kobe actually took 30,699 shots in his career. However, two shots seem to be missing from the official data at stats.nba.com. Multiple reports have found this: https://graphics.latimes.com/kobe-every-shot-ever/*

We can figure this out using this pattern:

**(# of possible ‘8-streaks’) = (total shots) – 7**

Note:

1. Longer streaks count for multiple 8 streaks.
2. Streaks can “wrap-around” from end of game to beginning of next game.

So we get the following:

**30,697 – 7 = 30,690 chances to 8-streak**

And we can get the expected number of 8-streaks below:

<img src="https://i.postimg.cc/bqrNMkCn/Perm-and-Comb-Kobe-Hot-Hand-Probability.png" title="Chance of 8-streak times chances to 8-streak equals 48.8 expected 8-streaks by chance" width = 300/>

#### Compare Expected to Observed

**Expected 8-streaks by chance = 48.8**

**Actual # of 8-streaks in career = 47** 

Very close! And the assumption of independence predicted well!

**Discussion Question:** Is this enough to disprove the “hot hand” theory? If yes, how so? If not, how else
could you test the theory?

<div class="alert alert-block alert-warning">

**Sample Response**

Not enough – A definitive statement cannot be made about the “hot hand” theory based on only one
example. We’d have to analyze multiple players and multiple eras in the league (along with other
generalizing factors) to come up with more comprehensive proof for or against the “hot hand” theory.

Furthermore, this example not only assumes the probably of each shot is independent, but it also
assumes the probabilities are identical – a stronger assumption that may be refutable.

several in a row. It’s possible that these hot and cold forces counterbalanced to create a similar number
of streaks (47) to what is predicted by the independence and equal probability assumptions alone (48.8).
