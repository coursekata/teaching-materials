## 5.0 Why does error matter?

<div class="teacher-note">

**Teacher Note** 
- Choose any of the discussion prompts below to help students grapple with the real implications of error. We want to connect **error → people → decisions** 
- Listen for: “Smaller residuals = more reliable advice,” “Big misses can cause harm,” “We should look for patterns in who gets mispredicted.”
- This is also an opportunity to think about correlation ≠ causation; models inform decisions, they don’t make them.
- Bridge to the next idea: if totals can hide **big misses**, we’ll need a measure that penalizes them more → (sets up squared error/SSE in 3.5). 

We included these questions in the next mini JNB if you want to save it til then. </div>

Discuss (Think–Pair–Share, then whole class):

- **For students & families:** If our **avg_cost → default_rate** model is off, how could that mislead someone choosing between a low-cost and a high-cost college?  
  *When would a big error about a school’s default rate really matter for a family’s decision?*

- **For counselors & schools:** How could a more accurate **avg_cost-based** model (smaller total error) help compare **similarly priced** schools more fairly?  
  *When might “price alone” be a weak signal?*

- **Fairness & impact:** If the **avg_cost** model consistently **underestimates** default rates for certain school types (e.g., for-profit) or price ranges, who could be harmed?  
  If it **overestimates** default for lower-cost schools, who might be discouraged unfairly?

- **Big misses vs. total:** Two **avg_cost** models can have similar total error, but one has a few **huge misses** (e.g., at very high costs). Which model would you trust more for advising students? Why?

- **Decision-making:** What other information (besides **avg_cost**) would you want before calling a college a “better investment”?  
  *(Examples: grad_rate, net_price after aid, median debt, med_alum_earnings, ownership/public vs. for-profit.)*
