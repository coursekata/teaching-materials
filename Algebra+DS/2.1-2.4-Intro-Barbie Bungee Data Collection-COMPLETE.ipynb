<div class="teacher-note">

<b>Goals of this section:</b> In this lesson, students engage in organizing and interpreting measurement data to begin reasoning about relationships between two quantities: rubber bands and drop height. It is designed to build foundational ideas that support later work with slope, y-intercept, and functional modeling.

<ul> <li> Students will practice <b>measuring and recording</b> variable quantities in a bivariate context. </li> <li> Students will explore how <b>changes in one quantity affect another</b>, reviewing slope as a rate of change and y-intercept as a starting value. </li> 
<li> Students will connect slope to real-world meaning — e.g., “each rubber band adds ~X cm to the drop.” </li> <li> Students will reflect on <b>variation and outliers</b> in physical data and consider sources of error or pattern. </li> 
</ul>
This activity anchors later lessons on graphing and regression by giving students an intuitive, physical feel for linear relationships and change across cases. It supports slope as a conceptual tool before reintroducing it symbolically or computationally.

- **Clear data entry system**:  
  - This notebook assumes you’ll provide a **Google Sheet** where students can enter their drop data.  
  - We recommend creating **separate tabs (sheets) for each class period**. That way, you can reuse the same link across periods while keeping data organized.
