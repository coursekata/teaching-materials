### 2.3 Better than total error 
The Sum of Squared Errors (SSE) is a way to measure how well a model fits data. Instead of just adding the residuals (which can cancel each other out), we square each residual first, then add them all together. Squaring makes every error positive and highlights larger mistakes more strongly.

### Try finding the SSE for both models below. 

```r
# write some code here 
```

```r
# sample response 
sum(empty_residual^2)
sum(my_function_residual^2) 
```

### Look at the graphs showing regular error and squared error. What does the SSE tell you about your models, and how does squaring the errors change the way you judge model fit?

<div class="teacher-note">

**Sample Responses:** 
- The SSE is always positive because we squared the residuals, so it doesn’t cancel out like total error.
    - Instead of adding positive and negative error, we are adding the *area* of the error squares
- my_function is better than the empty model because its SSE is smaller, which means the errors are smaller overall.  
- Squaring makes big errors stand out more, so it’s easier to see which model fits better. 
    - Imagine stacking the boxes -- with regular error, you would be stacking positive lines and negative lines.
