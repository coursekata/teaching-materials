## 0.0: Explore the `anime_tv` data frame.

[MyAnimeList (MAL)](https://en.wikipedia.org/wiki/MyAnimeList) is an anime and manga social networking and cataloging website. MAL provides a list-like system so users can organize and score preferred anime and manga. The site claims to have 4.4 million anime and 775,000 manga entries and records millions of visitors a month.

The `anime_tv` data frame contains 1,074 anime TV series with the following 15 variables:

- `name` anime title in English — extracted from MAL. Example: *Fullmetal Alchemist: Brotherhood*
- `title_japanese` title in Japanese (or in Chinese/Korean if the anime is from those countries)
- `type` anime type (e.g., Movie, Music, ONA-original net animation, OVA-original video animation, Special, TV); all anime in this data frame are TV shows
- `source` what the anime is based on (e.g., original story, manga, video game, music)
- `genre` genre of the anime (e.g., Action, Romance, Comedy)
- `studio` production studio that made the anime (e.g., MAPPA, Studio Ghibli)
- `episodes` number of episodes the anime has
- `decade` decade the anime began airing
- `rating` audience rating (e.g., PG-13, R+); similar to movie ratings
- `score` average rating on a 1 to 10 scale, based on MAL user reviews
- `scored_by` number of users who rated the anime; this data frame includes only anime with at least 25,000 ratings
- `rank` overall rank based on score — lower is better (e.g., #1 is top-ranked). See [Wikipedia](https://en.wikipedia.org/wiki/MyAnimeList) for details
- `popularity` popularity rank based on how many users added the anime to their list — lower means more popular
- `members` total number of MAL users who added the anime to their list
- `favorites` number of users who marked the anime as a favorite

These are a subset of data from the various rankings and scores available at [MyAnimeList.net](https://myanimelist.net). ## 6.0: Models that Perfectly Balance Residuals

The mean of `rank` was crucial to balancing the error. But so is the mean of `score`. You can use both the mean of `rank` and the mean of `score` to *perfectly balance* the residuals. 

<div class="alert alert-block alert-warning">

Just as the residuals around the mean of a variable add up to 0, the residuals around a linear function (the line we are using as a model) will add up to 0 if the line passes throught the ***point of means*** (`mean(X)`, `mean(Y)`).
