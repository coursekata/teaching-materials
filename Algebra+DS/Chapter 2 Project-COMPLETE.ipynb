### Teacher Resource: Talking About Homelessness with Students  

Because homelessness is a sensitive and often personal topic, it’s important to frame discussions in ways that are respectful, inclusive, and culturally responsive.  

**General Guidelines**  
- Remind students we are working with **state-level data**, not individual stories. Avoid stereotypes or assumptions about people who are unhoused.  
- Use **person-first language** (e.g., “people experiencing homelessness” rather than “the homeless”).  
- Emphasize that many factors influence homelessness (housing costs, health, systemic inequities) and that the data only shows correlations, not causes.  
- Prepare to gently redirect deficit-based or stigmatizing comments toward more structural/systemic explanations.  

**Resources for Teachers**  
- [Teaching Tolerance (Learning for Justice) – Classroom Resources](https://www.learningforjustice.org/classroom-resources)  
  Guidance on creating inclusive discussions around sensitive social topics.  
- [National Center for Homeless Education](https://nche.ed.gov/)  
  Information, data, and resources for understanding homelessness in schools and communities.  
- [Culturally Responsive Teaching Practices](https://www.csueastbay.edu/soe/faculty-and-staff/culturally-responsive-pedagogy.html)  
  Strategies for engaging students with empathy and cultural awareness.  
- [Teaching Students Experiencing Homelessness (National Coalition for the Homeless)](https://nationalhomeless.org/)  

**Discussion Moves**  
- Ask students to notice **patterns** and **questions**, rather than making judgments.  
- Encourage empathy by connecting homelessness to broader themes like housing, affordability, and access to resources.  
- Highlight that math and data can be tools for **understanding complex issues**, not for labeling people.

## Limitations

<div class="alert alert-block alert-info">

The goal of this section is to reflect on the limits of your model so your audience understands when to be cautious about your results.  

A good limitations section might include:  
- **When does your model not make sense?** 
    - Cite the domain and range here. 
    - Are there predictions that aren’t realistic?  
- **What might be missing?** Are there other variables or factors that could help explain the outcome that you didn’t include?  
- **What are the limits of the data?**  
  - The data is **state-level**, not about individual people. What does that mean for how we interpret results?  
  - The **PIT Count** is from one night in January. How might that affect the numbers?  
  - Some variables come from different years or different sources. How might that matter?  

Remember: pointing out limitations doesn’t weaken your project — it shows you are thinking critically about what your model can and cannot tell us.
