<b>Teacher note:</b>  
Students may raise two broad types of issues — both are worth exploring:

- **Who are we measuring? (Quality of Sample)**  
  Students might question whether this set of cases represents people more generally.  
  This is a good time to introduce the term **convenience sample** — a sample that’s easy to collect (like people in our class), but may not reflect the larger population.  
  You can contrast this with a **random sample**, which is designed to better represent a broader group.

- **How good are our measurements? (Quality of Measurement)**  
  Students might also notice that people aren’t great at estimating zombie survival time — especially when someone writes something like 1,000,000 days.  
  This opens up a conversation about how well a variable captures what we want to measure.
