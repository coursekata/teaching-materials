#### Data from counties in the United States that have had UFO sightings.

For the purpose of this data investigation, you can believe in these UFO sightings or be a skeptic! The data frame `ufo` contains data from reported UFO sightings in the Pacific Region: California and Hawaii. It contains these variables:

- `county_fips` FIPS is a four-digit Federal Information Processing Series (FIPS) code unique to the county
- `county_name` name of county
- `state_name` the state that the county is in
- `timezone` the timezone of the county
- `sightings` number of reported sightings in the county
- `avg_encounter` average length of the encounter in seconds
- `light_count` how many of the reported sightings mentioned a light
- `round_count` how many of the reported sightings mentioned a round, circular, or spherical object
- `area` land area of the county in square miles
- `population_k` the population of the county (from Census data) in thousands (K)
- `density` the population (number of individuals) per square mile
