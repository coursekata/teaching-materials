The two major political parties know that if you win a state (even by only a little bit), you get **ALL** of its electoral votes. There are states where if the political campaigns make the effort to get a few more votes, they can get all of the electoral college votes. These are often called the *swing states* (also called *battleground states*).

#### Variables

- `state`: Name of state
- `state_code`: Two-letter abbreviation of state name
- `r_votes16`: Number of votes for Republican candidate in year 2016
- `d_votes16`: Number of votes for Democratic candidate in year 2016	
- `r_votes20`: Number of votes for Republican candidate in year 2020	
- `d_votes20`: Number of votes for Democratic candidate in year 2020
- `pop_size`: Population size of state
- `house_seats`: Number of house representatives 
- `senate_seats`: Number of senators (2 Senate seats per state)
- `electoral_votes`: Number of electoral votes (towards the Electoral College) 
- `campaign_events20`: Number of campaign events hosted by Presidential candidates in the state between Aug 28 - Nov 3, 2020
- `facebook_ad24`: Amount of weekly spending on political ads by any PAC or superPAC on facebook in dollars per person (i.e., .09 means 9 cents per person)
- `facebook_ad_total24`: Total amount of weekly spending on political ads by any PAC or superPAC on facebook in dollars

Now let's create a new column using `mutate()` and see which states had the closest races.

We’ll use one possible definition of swingyness: the difference between Republican and Democratic votes in 2016. This gives us a number that tells us how close the election was. 

#### `vote_diff16` = `r_votes16` - `d_votes16`

```r
# mutate the elections dataframe 
elections <- elections %>%
  mutate(vote_diff16 = r_votes16 - d_votes16)

# let's look at the data frame to see what's changed
head(elections)
```
