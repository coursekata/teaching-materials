### Teacher Note: Avoiding Deficit Framings in Global Data

When working with global datasets (e.g., GDP, life expectancy, education), students may notice that countries in Africa or the Global South often appear lower on certain indicators. This can unintentionally reinforce **deficit-oriented narratives**, framing entire regions as “behind,” “poor,” or “less developed.”

#### Why it matters
- Data doesn’t tell the whole story — it reflects **historical power imbalances**, **colonial legacies**, and **global economic systems**.
- Framing a country’s statistics without context can unintentionally send the message that people or cultures are to blame, rather than systems.

#### Strategies to use
- **Shift the conversation** from “why are they behind?” to “what systems and histories have led to these outcomes?”
- **Highlight resilience and strengths** alongside challenges (e.g., innovations in public health, education access improvements).
- **Challenge language** that reinforces hierarchy (e.g., instead of “third world,” use “Global South” or “low-income countries,” and explain why).

## About `World`

We'll be looking at a data set called `World`, with data on countries from the Happy Planet Index project, updated with some historical height data (from [Our World in Data](https://ourworldindata.org/human-height)), drinking data (collected by the World Health Organization featured in [fivethirtyeight](https://fivethirtyeight.com/features/dear-mona-followup-where-do-people-drink-the-most-beer-wine-and-spirits/)), population and land characteristics, and vaccination data (from March 2023). 

It has 130 countries and the following variables:

- `Country` Name of country
- `Region` One of 5 UN defined regions: Africa, Americas, Asia, Europe, Oceania 
- `Code` Three-letter country codes defined by the International Organization for Standardization [(ISO)](https://www.iso.org/iso-3166-country-codes.html) to represent countries in a way that avoids errors since a country’s name changes depending on the language being used.
- `LifeExpectancy` Average life expectancy (in years)
- `GirlsH1900` The average of 18-year-old girls heights in 1900 (in cm)
- `GirlsH1980`  The average of 18-year-old girls heights in 1980 (in cm)
- `Happiness` Score on a 0-10 scale for average level of happiness (10 being happiest)
- `GDPperCapita` Gross Domestic Product (per capita)
- `FertRate` The average number of children that will be born to a woman over her lifetime
- `PeopleVacc` total number of people vaccinated in the country
- `PeopleVacc_per100` ratio (in percent) between the total number of people vaccinated and total population in the country
- `Population2010` Population (in millions) in 2010
- `Population2020` Population (in millions) in 2020
- `WineServ` Average wine consumption per capita for those age 15 and over per week (collected by WHO)

**Sample Response:**

1. **Hypothesis (4.1):**  
   I think countries with higher happiness scores will have longer life expectancy.

2. **Graph (4.2):**

3. **Linear Model (4.3):**  
   (See model below.)  
   $LifeExpectancy = 30 + 6.2 \times Happiness$

4. **Interpret Parameters (4.4):**  
   In this equation, the y-intercept ($b_0 = 30$) means that if a country's happiness score were 0 (which isn’t realistic), the predicted life expectancy would be 30 years.  
   The slope ($b_1 = 6.2$) means that for every 1-point increase in happiness, the life expectancy is predicted to increase by about 6.2 years.

5. **Model Fit (4.5):**  
   A linear model seems like a reasonable fit here — the dots follow a pretty clear upward trend. But we should be cautious about drawing conclusions. The graph shows correlation, not causation. Also, these are averages for countries, not individuals. So we can’t say happiness directly causes longer life (*directionality problem*), or rule out a third factor causing both (*third-variable problem*).

</div>

```r
gf_point(LifeExpectancy ~ Happiness, data = World, color = "purple") %>%
  gf_abline(intercept = 30, slope = 6.2)
```
