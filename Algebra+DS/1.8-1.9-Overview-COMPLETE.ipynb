These are often called the *swing states* (also called *battleground states*).

Now let's create a new column using `mutate()` and see which states had the closest races.

We’ll use one possible definition of swingyness: the difference between Republican and Democratic votes in 2016. This gives us a number that tells us how close the election was.

- Swingy states are in the middle when we don’t use abs() — because they’re the ones with vote differences close to zero.

<b>Teacher Note:</b>
- When vote_diff is calculated without abs(), negative values indicate states where Democrats led, and positive values mean Republicans led. The most competitive (swingy) states are those where the value is close to zero — so they appear in the middle of the list. 
- Encourage students to interpret what high, low, and near-zero values mean in context, and consider letting them compare sorted versions with and without abs().
