<b>Goals of this section:</b>   In this lesson, students extend their understanding of error by exploring <b>Mean Squared Error (MSE)</b>, <b>Root Mean Squared Error (RMSE)</b>, and <b>Proportional Reduction in Error (PRE)</b>. These measures provide new ways to interpret model fit, making error comparisons easier to understand and communicate. Students see that while SSE depends on the scale of the data, MSE and RMSE adjust for that, and PRE expresses model improvement as a proportion of total error.  

<ul> 
  <li>Students will calculate the <b>Mean Squared Error (MSE)</b> as the average squared residual, understanding how it adjusts for sample size.</li>
  <li>Students will calculate the <b>Root Mean Squared Error (RMSE)</b> and interpret it as the typical prediction error in the original units of the outcome variable.</li>
  <li>Students will calculate the <b>Proportional Reduction in Error (PRE)</b> to see how much of the total error is reduced by the regression model compared to the empty model.</li>
  <li>Students will interpret MSE, RMSE, and PRE in context, comparing what each reveals about model fit.</li>
</ul>

This lesson builds students’ toolkit for evaluating regression models by introducing error measures that are both interpretable and comparable. It prepares them to connect regression analysis to the concept of variance explained and to judge model quality more reliably.

Sum of squares is really helpful for finding the best-fitting model, but it’s a weird number to actually interpret and SSE gets bigger and bigger as data sets get bigger. If there is more data, then there will be more total square error. 

In answer to these difficulties, statisticians have devised some measures based on SSE, but more interpretable. These are **Mean Square Error (MSE)** and **Root Mean Square Error (RMSE)**.

<div class="guided-notes">

### 3.2 Let's define the different measures of error in our guided notes. 

</div>

### 3.3 Calculate these measures of error for your best-fitting model.

## 4.0 How much does your model *reduce* error?

Another way to evaluate our model is to compare it to the empty model -- if you were trying to predict `loveable` without a predictor variable. 

### 4.1 Create the empty model for your predictor variable and calculate the SSE. 

````r
# create the empty model 
empty_model <- 

# calculate the SSE for the empty model 
````

````r
#sample code 

# create the empty model 
empty_model <- function(X){mean(Y)}

# calculate the SSE for the empty model 
sse(Y, empty_model(X))
````

<div class="alert alert-block alert-warning">
 
 #### This is the *total error*, or the error if we predicted `loveable` without using any predictor variable.

<div class="teacher-note">

**Teacher Note:** You could use the SST notation here and have students store the SSE of the empty model.
    <code>sst <- sse(Y, empty_model(X)) </code>
    <code>sse <- sse(Y, my_function(X)) </code>
Then you can use SST and SSE instead of the full codeline in the next steps. 

### 4.2 How much smaller is your model's error compared to the empty model?

````r
# write some code here 
````

````r
# sample code 

sse(Y, empty_model(X)) - sse(Y, my_function(X))

# or using SSE and SST
sst <- sse(Y, empty_model(X))
sse <- sse(Y, my_function(X))

(sst - sse)
````

<div class="teacher-note">
 
 You just calculated the difference between the total error (SST) and your model’s error (SSE). 
 #### This tells use how much your mode *reduced* the total error. 

### 4.3 How much of the total error does this reduction represent?

````r
# write some code here 
````

````r
# sample code 

(sse(Y, empty_model(X)) - sse(Y, my_function(X)))/sse(Y, empty_model(X))

# or using SSE and SST

(sst - sse)/sst
````
