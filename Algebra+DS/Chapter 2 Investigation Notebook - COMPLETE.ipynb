This investigation is also an opportunity to challenge students to think carefully about **correlation vs. causation**. Some relationships may look strong but are unlikely to be causal (e.g., a positive correlation between state-level income and homelessness counts). These examples provide powerful discussion opportunities: states with higher incomes may also have higher housing costs, which relates to homelessness — but this does *not* mean individuals with higher incomes are more likely to be homeless.

## 4.0 Using your model

### 4.1 If you haven't already, program your model as a custom function in R. 

Then use `gf_function()` to put your line on your scatterplot.

```r
#write some code here 
```

```r
# sample response
my_function <- function(X){145 + -1.5*X}

gf_point(Obesity ~ PhysicalActivity, data = States) %>%
  gf_function(my_function)
```

### 4.2 Use your custom function to generate a few predictions. 
Where will those predictions appear on our scatterplot? **Add the predictions to your scatterplot.**

```r
# sample response
my_function(70)
my_function(80)
my_function(100)

gf_point(Obesity ~ PhysicalActivity, data = States) %>%
  gf_function(my_function) %>%
  gf_point(my_function(70) ~ 70, color = "magenta")%>%
  gf_point(my_function(80) ~ 80, color = "magenta")%>%
  gf_point(my_function(100) ~ 100, color = "magenta")
```

**Sample response:**

`my_function(70)` is 145 + (-1.5)*70 --> 40 = f(70); This seems reasonable compared to the scatterplot.

**Teacher Note:** Some groups may test predictor values that are within the domain of the data, while others may try values outside the domain.
- Use this as a discussion point: Which predictions seem reasonable, and which don’t make sense? Why?
- Highlight that models are strongest inside the range of observed data. Predictions outside this range (extrapolations) can be misleading, but they still help us see the limits of linear models.
- If time allows, ask groups to share one “reasonable” prediction and one “unreasonable” prediction, and compare how they justified each.
