## 1.0: What do you know about semiconductor chips?

There is a lot of mystery around AI but much of the advances in AI are made possible by semiconductor chips. These tiny little devices have a broad range of applications. According to the [Semiconductor Industry Association](https://www.semiconductors.org/semiconductors-101/what-is-a-semiconductor/#:~:text=Semiconductors%20are%20an%20essential%20component,energy%2C%20and%20countless%20other%20applications.): "Semiconductors are the Brains of Modern Electronics. They are an essential component of electronic devices, enabling advances in communications, computing, healthcare, military systems, transportation, clean energy, and countless other applications." We might want to learn more about them! 

The data frame called `chip_dat` contains data from https://chip-dataset.vercel.app/

Here are some of the variables in the data frame:
- `product` name of the product 
- `type` there are CPU or GPU type chips 
- `release_date` the date this product was released
- `year` the year of release formatted as a two-digit year (e.g., 11 means the year 2011); all chips in data set were released after 2000
- `process_size_nm` The size of individual processors (also called transistors) in nanometers (in nm; for scale, a DNA double helix is about 10 nm)
- `tdp_w` thermal design power (TDP), sometimes called thermal design point, the maximum amount of heat generated by a chip (in watts) 
- `die_size_mm2` the die is the actual bare chip; die size is the length x width (in mm squared)
- `process_mill` how many processors are on a chip (in millions)
- `freq_mhz` clock frequency rates are a measure of a chip's speed (in MegaHertz, 1 million cycles per second); frequency rate is the speed at which the clock generator of a processor can generate pulses, which are used to synchronize the operations of its components.
- `foundry` a semiconductor manufacturer that makes chips for other companies
- `vendor` the company that sells the chip
- `FP32_GFLOPS` FLOPS is floating point operations per second, a measure of computer performance. GFLOPS is gigaFLOPS. FP32 indicates half-precision floating point format that uses 32 bits (which is faster and less precise than 64 bits). FP32 is also more common than FP64.  
- `FP64_GFLOPS` See `FP32_GFLOPS`
