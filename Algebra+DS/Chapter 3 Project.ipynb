## Limitations

<div class="alert alert-block alert-info">

The goal of this section is to reflect on the **limits of your model**, so your audience understands when to be cautious about your results.

A strong limitations section might include:

- **When does your model not make sense?**  
- Think about the **domain and range** of your model.  
- Are there predictions that don’t seem realistic (e.g., negative earnings, extremely high values)?  

- **What might be missing?**  
   - Are there other factors that could help explain the outcome that you didn’t include?  
     (For example: student major, location, selectivity, or job market conditions.)  

- **What are the limits of the data?**  
   - The data is **institution-level**, not about individual students. How does that affect how we interpret your model?  
   - Some colleges may report data from different years or use different definitions. Could that affect consistency?  
   - What do you notice about missing data or unusual values?

>  Remember: Pointing out limitations doesn’t weaken your project — it shows that you’re thinking like a data scientist. Good models help us make better decisions, but they don’t tell the whole story.
