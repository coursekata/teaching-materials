### 1.2 Definitions of all the variables

The data set includes the following:
- `name`: The name of the Pokémon  
- `pokedex_number`: Its number in the National Pokédex  
- `type1`: A Pokémon’s main type (like Fire, Water, or Grass), which affects which moves it’s strong or weak against  
- `type2`: A second type (if it has one)  
- `num_type`: The number of types the Pokémon has (1 or 2)  
- `height`: The Pokémon’s height in meters (m)  
- `weight`: The Pokémon’s weight in kilograms (kg)  
- `capture_rate`: How easy or hard it is to catch the Pokémon (higher = easier)  
- `total`: The sum of all the stats listed below — a rough measure of how strong it is overall  
- `hit_points`: How much health it has (how much damage it can take before fainting)  
- `attack`: How strong it is with physical moves (like punches or tackles)  
- `defense`: How well it can block physical attacks  
- `sp_attack`: How strong it is with special moves (like Fire Blast or Thunderbolt)  
- `sp_defense`: How well it can block special moves  
- `speed`: How fast it is — faster Pokémon usually go first in battle  
- `generation`: Which generation of Pokémon games it was introduced in  
- `legendary`: Whether it's a Legendary Pokémon (super rare and usually very powerful)  
- `category`: A broad group for its type (like Elemental, Physical, or Mystic)  
- `magic`: Whether the Pokémon uses more “magical” or mysterious types and moves (True/False)  
- `friendship`: How naturally friendly the Pokémon is with its trainer — this affects how quickly it bonds in the games  
- `recommended_ball`: A suggestion for the kind of Poké Ball that would work well when trying to catch it

This dataset was sourced from [Kaggle](https://www.kaggle.com/datasets/rounakbanik/pokemon?resource=download).

### 2.4 Definition of outcome and explanatory variables

Using past data to make predictions—like predicting how powerful a pokemon is—is something statisticians and data analysts do all the time. To do this well, it's important to understand the two roles that variables can play.

The **outcome variable** (or *output*) is what you're trying to predict. There's usually just one. 

**Explanatory variables** (or *inputs*) are the other things you know about each observation that might help you make a better prediction of the outcome.

### 2.6 Description of word equations 

Your hypothesis about how an explanatory variable might help to predict an observation's value on an outcome variable can be expressed as a word equation. You can think of word equations as *informal models* of the world. Word equations take the form:

- **Outcome Variable = Explanatory Variable + Other Stuff**

For example, one hypothesis might be that a Pokémon's weight might be related to their total strength. We could express this idea as:

- **total = weight + Other Stuff**
