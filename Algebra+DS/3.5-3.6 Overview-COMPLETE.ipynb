### 3.3 The EMPTY MODEL

This model is special! This model, `score` = `mean` + `error` is called the ***empty model***. Because the empty model uses the mean of $Y$, the total error is 0 (or very very close). Residuals are balanced at the mean of a variable.

## 6.0: Models that Perfectly Balance Residuals

The mean of `rank` was crucial to balancing the error. But so is the mean of `score`. You can use both the mean of `rank` and the mean of `score` to *perfectly balance* the residuals. 

<div class="alert alert-block alert-warning">

Just as the residuals around the mean of a variable add up to 0, the residuals around a linear function (the line we are using as a model) will add up to 0 if the line passes throught the ***point of means*** (`mean(X)`, `mean(Y)`). 

<div class="guided-notes">

### 6.1 What's special about models that have perfectly balanced residuals?

### 6.2: Create a custom function for a line that goes through this special point. 

We want to fill in this equation: $Y = b_0 + b_1X$

If we know 3 of the four pieces, we can solve for the final piece. We can plug in `mean(X)` for $X$ and `mean(Y)` for $Y$ because this line has to go through the point of means. 

> This means we need to find a $b_0$ and $b_1$ for this line.

<div class="teacher-note">

**Teacher Note:** 
- Having numbers for $b_1$, an example $X$ and an example $Y$, we can solve for $b_0$. You can show students how to code it up! 
- Students' first instinct will be to actually plug in the numbers (e.g., 20,000, the mean of Y (1154.92) and the mean of X (7.87)). They can do that first but without all the decimal values, it won't exactly go through the point all the time.


```r
# sample response

# put in your estimate for b1
b1 <- -8000/2.5
# here, solve for y-intercept
b0 <- mean(Y) - b1*mean(X)

#print out your b1 and b0 
print("slope") 
b1

print("intercept") 
b0


# this will create a function from this set of b0 and b1
my_balanced_function <- function(X){b0 + b1*X}

# adding to the graph
gf_point(rank ~ score, data = anime_tv) %>%
  gf_point(mean(Y) ~ mean(X), color = "red") %>%
  gf_function(my_balanced_function)
```
