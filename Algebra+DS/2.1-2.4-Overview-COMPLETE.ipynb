## <font color=deeppink>3.0: Modeling Drop Distance</font>

### 3.1 From points to a line: `gf_abline`
In 2.0, you made individual predictions by plotting dots representing different numbers of rubber bands.

But instead of placing each dot one at a time, what if we could express all our predictions with a single line?

A line helps us quickly model how Barbie’s drop distance changes as the number of rubber bands increases. Just like before, this line represents a statistical model — one that generates a prediction for every possible number of bands.

### Modify the y-intercept and slope below to create a line of predictions.
Try to match the general trend of the data, even if your line doesn’t hit every point.

```r
#edit this code

gf_point(distance ~ bands, data = bungee, color = "hotpink") %>%
    gf_abline(intercept = 100, slope = 0)
```

<div class="teacher-note">

**Teacher Note:**  
- Encourage students to interpret the slope and intercept in terms of Barbie’s fall: the slope is how much farther she drops with each added band; the intercept represents the starting drop distance with zero bands. This is a great opportunity to connect to `y = mx + b`.
- Try doing one demo first (e.g. slope = 20, intercept = 18) to help students see what each number controls. Then give time to adjust on their own.
- Then let students work through 3.3 with a partner or group. 

</div>

```r
#COMPLETE

gf_point(distance ~ bands, data = bungee) %>%
    gf_abline(intercept = 18, slope = 20, color = "hotpink")
```

<div class="alert alert-block alert-warning">

In algebra, you may have learned to write a line as:

> \( $y = mx + b $\)

But in statistics and data science, it’s common to use **GLM notation** — short for **General Linear Model**. We will use this notation because it is more often used by professionals in the statistics and data science world, and because it will make it easier later when you learn about other models (other than a straight line).

We write the same equation as:

> \( $Y = b_0 + b_1X$ \)


Why the change?

- The order is reversed: \( b + mx \) instead of \( mx + b \) 
- The symbols are different:  
  - \( $b_0$ \) is the **intercept** (like your starting point)
  - \( $b_1$ \) is the **slope** (how much Y changes for each unit of X)
</div>
