<div class="teacher-note">

<b>Goals of this section:</b> In this lesson, students will practice organizing and exploring data by learning three foundational data wrangling tools: <code>select()</code>, <code>filter()</code>, and <code>arrange()</code>. These tools allow us to focus on relevant variables, narrow our data to specific cases, and sort it to find meaningful patterns — essential skills in both data science and algebraic modeling.

  <ul>
    <li> Students will learn to <b>select</b> specific columns (variables) from a dataset using <code>select()</code>.
    <li> Students will learn to <b>filter</b> for specific rows (cases) that meet a condition using <code>filter()</code>.
    <li> Students will learn to <b>arrange</b> rows by increasing or decreasing values of a variable using <code>arrange()</code>.
    <li> Students will connect these operations to real-world reasoning — such as identifying swing states or sorting by voter turnout — and begin to ask investigative questions that data can answer.
  </ul>

  This section builds fluency with core data wrangling functions in R, setting students up for deeper exploration, modeling, and visualization in later lessons.
</div>

### The Dataset

**Description:** The `mini_elections` data frame contains the data printed on your paper data frames. The dataset has tidy organization, with one row per state.

#### Variables

- `state`: Name of the state  
- `r_votes16`: Number of votes for the Republican presidential candidate in 2016  
- `d_votes16`: Number of votes for the Democratic presidential candidate in 2016  
- `total_votes16`: Total number of votes cast in the 2016 presidential election (Republican + Democratic votes)  
- `house_seats`: Number of representatives the state sends to the U.S. House of Representatives. This number depends on the state's population.  
- `senate_seats`: Number of U.S. Senators from the state. Every state has 2 senators, no matter how big or small.  
- `electoral_votes`: Total number of votes the state gets in the Electoral College (used to choose the President). This is equal to the number of House seats + 2 Senators.

#### Data Sources

- **Election Results Data:** MIT Election Data and Science Lab, 2017, ["U.S. President 1976–2020"](https://doi.org/10.7910/DVN/42MVDX), Harvard Dataverse, V6  
- **Electoral College Info:** [archives.gov](https://www.archives.gov/electoral-college/allocation)
